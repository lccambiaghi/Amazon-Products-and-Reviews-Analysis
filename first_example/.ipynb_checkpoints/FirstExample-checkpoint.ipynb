```python
import pandas as pd
import json
import time
import datetime

start = time.time()
luca_path = "/Volumes/HD_Journaled/Dataset/reviews_Kindle_Store_5"
riky_path = "C:\\dtu\\ctbd\\amazon_dataset\\Kindle_Store_5.json"
path = riky_path
with open(path) as data: #This JSON file is around ~800MB
    lines = [json.loads(line) for line in data.readlines()]

df = pd.DataFrame([elm.values() for elm in lines],columns=lines[0].keys())
end = time.time()
print 'Dataset importation time:',datetime.timedelta(seconds=(end-start))
```

```python
df
```

```python
from pandas import DataFrame
df_grouped_obj = df.groupby('asin',as_index=False).size()
df_count = DataFrame({'num_of_reviews':df_grouped_obj}).reset_index()
df_count = df_count.sort_values(by='num_of_reviews',ascending=False)
print 'Top reviewd products:'
df_count.where(df_count['num_of_reviews']>5).dropna().head()
```
